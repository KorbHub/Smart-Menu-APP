```python
import requests

# Define the endpoint and parameters for the API request
endpoint = 'https://api.edamam.com/api/nutrition-data'
params = {
    'app_id': '742dfbb0',
    'app_key': '86d612a8156f10f9332196b6f9a4d208',
    'ingr': 'Beef Curry'
}

# Send a GET request to the API and parse the JSON response
response = requests.get(endpoint, params=params)
data = response.json()

# Extract the nutrient and allergy information from the response
Health_label = data['healthLabels']
Dietary_choices = [x for x in  Health_label if not x.endswith("_FREE")]
Allergies = [x for x in Health_label if x.endswith("_FREE")]
# Display the information to the user or use it for further processing
print(Health_label,Dietary_choices,Allergies)
```
